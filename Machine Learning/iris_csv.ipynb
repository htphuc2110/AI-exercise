```python
import pandas as pd
url = 'https://raw.githubusercontent.com/htphuc2110/File-/main/iris.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()
```

```python
correlation = data.corr()
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation, annot=True)
```

### ***Hi quy Bc 1***

```python
data.plot(kind='scatter',x='petal.width',y='petal.length',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['petal.width']),y=data['petal.length'])
print('Gia tri Intercept la: ',model.intercept_)
print('Gia tri Coeffient la: ',model.coef_)
```

```python
model.score(X=pd.DataFrame(data['petal.width']),y=data['petal.length'])
```

```python
data.plot(kind='scatter',x='petal.width',y='petal.length',figsize=(9,9),color='black')
y_pred = model.predict(X=pd.DataFrame(data['petal.width']))
plt.plot(data['petal.width'],y_pred,color='blue')
```

### ***Hi quy bc 2***

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['petal.width'],data['petal.width']**2]).T
y = data['petal.length']
model.fit(X,y)
print('He so Intercept la : ',model.intercept_)
print('He so coeffient la :',model.coef_)
print('Do chinh xac cua mo hinh :',model.score(X,y))
```

```python
import numpy as np
range = np.arange(0.0,2.6,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred  = model.predict(X_range)
data.plot(kind='scatter',x='petal.width',y='petal.length',figsize=(9,9),color='black',xlim=(-0.05))
plt.plot(range, y_pred, color='blue')
```
